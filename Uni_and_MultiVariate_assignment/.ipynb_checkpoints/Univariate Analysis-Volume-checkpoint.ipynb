```python
import pandas as pd 
import numpy as np
import seaborn as sns
variable = 'volume'
```

```python
data = pd.read_csv('diamonds.csv')
data.head()
```

```python
# We must Firt Clean the data as it may be dirty
```

```python
# Check for null values in each column
null_values = data.isnull().sum()

# Display the count of null values for each column
print(null_values)
```

```python
# The Data does not contain any empty values
```

```python
import matplotlib.pyplot as plt
```

```python
data['volume'] = data['x'] * data['y'] * data['z']
# new colum called volume added which is the space taken by a diamond
data.head()
```

```python
# we shall look into price as the variable we are studying
summary_stats = data[variable].describe()
print(summary_stats)
```

```python
sns.histplot(data=data, x=variable, bins=30)
plt.show()
# the graph shows an abnormal trend , the fact that the x axis has 4000 on it implies an outlier value
```

```python
sns.violinplot(data=data, x=variable)
# the violin plot also implies an outlier
```

```python
sns.kdeplot(data=data, x= variable, common_norm = False)
```

```python
sns.kdeplot(data[variable], fill=True , common_norm=False)
```
