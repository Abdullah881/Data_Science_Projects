```python
import pandas as pd 
import numpy as np
import seaborn as sns

data = pd.read_csv('diamonds.csv')
data.head()
```

```python
# we are going to be comparing the price and carat variable
# we know from our previous univariate analysis that the price column contains no outliers or zero values
# we know that price and carat both do not contain nan values
```

```python
sns.boxplot(data=data, x='carat')
# the box plot allows us to see posible outliers , there do appear to be some possible outliers
# this may be relevent as price has no outliers
```

```python
(data[data['carat'] == 0])
# the carat column does not contain any zero values, whcih is ideal 
```

```python
sns.violinplot(data=data, x='carat')
# the box plot allows us to see posible outliers , there do appear to be some possible outliers
# this may be relevent as price has no outliers
```

```python
sns.boxplot(data=data, x = 'price')
# no outliers in price
```

```python
sns.jointplot(data=data, x='carat', y='price', kind='scatter')

# here we can see the outliers , price increases with carat, which makes sense as carat is a measure of weight 
# but there do seem to be some examples of diamonds of lower carat being priced more than heavier diamonds
# this does not mean an error as other factor such as clarity and cut are important factor of diamond price
```

```python
# Calculate the correlation coefficient
correlation = data['carat'].corr(data['price'])
print(f"Correlation coefficient between 'carat' and 'price': {correlation}")
```

```python
# given that correlation coefficient can be between -1<c<1 , a value this close to one implies that there is strong
# correlation between carat and price 
```

```python
# Grouping by 'color' and calculating the average price and carat for each group
average_stats_color = data.groupby('color').agg({'price': 'mean', 'carat': 'mean'})
print(average_stats_color)
# this shows how the price increases by colour but this is not a great measure as the avg carat size is also greater per colour 
# and we have proved carat increases price 
```

```python

# Calculate price per carat
average_stats_color['price_per_carat'] = average_stats_color['price'] / average_stats_color['carat']
print(average_stats_color)
```

```python
# it seems as color improves price increases but there is a fluctuation at the end D diamonds should be the most expensive
# how ever other factors may also be at play , e,g in our set all our D diamonds may be of lower cut, or smaller
```

```python
# Grouping by 'cut' and calculating the average price and carat for each group
average_stats_cut = data.groupby('cut').agg({'price': 'mean', 'carat': 'mean'})

# Calculate price per carat
average_stats_cut['price_per_carat'] = average_stats_cut['price'] / average_stats_cut['carat']
print(average_stats_cut)
```

```python
# this trend shows a consistent change as diamond quality improves so does price , it should be noted 
# premium is better than very good 
```
