```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
textmining = pd.read_csv('trainTweets.csv')
```

```python
textmining
```

```python
# Adding a colum called 'Word_count'
# function which counts how many words are in the column tweet
textmining['word_count'] = textmining['tweet'].apply(lambda x: len(str(x).split(" ")))
```

```python
textmining
```

```python
# Adding a colum called 'char_count'
# function which counts how many characters are in the column tweet, this includes blank spaces
textmining['char_count'] = textmining['tweet'].apply(lambda x: len(str(x)))
textmining
```

```python
textmining['hashtags'] = textmining['tweet'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
textmining.head()
```

```python
textmining['tweet'] = textmining['tweet'].apply(lambda x: " ".join(x.lower() for x in x.split()))
textmining.head()
# textmining['tweet'] = textmining['tweet'].str.lower()
# textmining.head()
```

```python
textmining['tweet'] = textmining['tweet'].str.replace('[^\w\s]','')
textmining.head()
```

```python
textmining['tweet'] = textmining['tweet'].str.replace('[^A-Za-z0-9\s]', '')
textmining.head()
```
