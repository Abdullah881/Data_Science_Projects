### Create a time series that compares life expectancy for Asia and Africa for the year 1952.


```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
# import Libraries 
```

```python
data = pd.read_csv('countries.csv')
data.head()
# import Data
```

```python
asia_1952 = data[(data['continent'] == 'Asia') & (data['year'] == 1952)]
asia_1952
# Create New dataframe that contains all asia data from 1952
```

```python
africa_1952 = data[(data['continent'] == 'Africa') & (data['year'] == 1952)]
africa_1952
# Create New dataframe that contains all Africa data from 1952
```

```python

# Create separate histograms for Asia and Africa

plt.subplot(2,1,1)
# Histogram for Asia
plt.hist(asia_1952['lifeExpectancy'], bins=15, color='blue', edgecolor='black')
plt.title('Life Expectancy in Asia (1952)')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.xlim(25, 70)  

plt.subplot(2,1,2)
# Histogram for Africa
plt.hist(africa_1952['lifeExpectancy'], bins=15, color='green', edgecolor='black')
plt.title('Life Expectancy in Africa (1952)')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.xlim(25, 70)  

plt.tight_layout()
plt.show()
```
